# **Training**

> Do not run below two cells as it would take time.

> The trained model is already saved in the given folder





```

transform_img = transforms.Compose([
    transforms.ToPILImage(),
    transforms.Resize(800),
    transforms.RandomHorizontalFlip(0.5),
    transforms.ToTensor(),
    utils.normalize_transform(),
])

dataset = core.Dataset('train.csv', 'images/', transform=transform_img)


image, target = dataset[2]


visualize.show_labeled_image(image, target['boxes'], target['labels'])
```

```

val_dataset = core.Dataset('val.csv', 'images/')

loader = core.DataLoader(dataset, batch_size=2, shuffle=True)

model = core.Model(['Number'])

losses = model.fit(loader, val_dataset, epochs=20, verbose=True)

plt.plot(losses)
plt.show()
```

```
model.save('model_weights.pth') 
```

### **Add images for prediction text**

> Give image path 



```
image_path='images/img2.jpeg' # Give Image path for masking !!
image=utils.read_image(image_path)
labels, boxes, scores = model.predict(image)
```

```
j=0
x1,y1,x2,y2=[],[],[],[]
for i in scores:
  
  if i>0.9:
    x1.append(int(boxes[j][0].round()))
    y1.append(int(boxes[j][1].round()))
    x2.append(int(boxes[j][2].round()))
    y2.append(int(boxes[j][3].round()))
    j+=1
  else:
    break
```
